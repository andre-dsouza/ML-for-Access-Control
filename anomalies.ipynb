```python

#import needed PyOD models

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
#%matplotlib inline
import matplotlib.font_manager
import pandas as pd


# Import models
from pyod.models.abod import ABOD
from pyod.models.cblof import CBLOF
from pyod.models.feature_bagging import FeatureBagging
from pyod.models.hbos import HBOS
from pyod.models.iforest import IForest
from pyod.models.knn import KNN
from pyod.models.lof import LOF

from pyod.utils.data import get_outliers_inliers

```

```python
#list of resource usage feature pairings
cols = list(usage)
for i in range(len(cols)):
    cola = cols[i]
    for j in range(i+1, len(cols)):
        colb  = cols[j]
        #print(cola, colb)
        print(cola.replace('resource_usage_','')+ ' & '+ colb.replace('resource_usage_',''))
```

```python
#ensemble outputs of isolation forest, HBOD and CBLOF

outliers = pd.DataFrame()
sums = pd.DataFrame()
legA = 'A'
leg = pd.DataFrame(columns = ['Key', 'Feature A', 'Feature B', 'Classifier', 'n_outliers', 'n_inliers', '% split'])
cols = list(usage)
for i in range(len(cols)):
    cola = cols[i]
    for j in range(i+1, len(cols)):
        colb  = cols[j]
        



        #data prep
        df2 = usage[[cola, colb]].copy()

        X1 = df2[cola].values.reshape(-1,1)
        X2 = df2[colb].values.reshape(-1,1)

        X = np.concatenate((X1,X2),axis=1)


        #model defining
        random_state = np.random.RandomState(42)
        outliers_fraction = 0.05
        # Define seven outlier detection tools to be compared
        # classifiers = {
        #         'Angle-based Outlier Detector (ABOD)': ABOD(contamination=outliers_fraction),
        #         'Cluster-based Local Outlier Factor (CBLOF)':CBLOF(contamination=outliers_fraction,check_estimator=False, random_state=random_state),
        #         'Feature Bagging':FeatureBagging(LOF(n_neighbors=35),contamination=outliers_fraction,check_estimator=False,random_state=random_state),
        #         'Histogram-base Outlier Detection (HBOS)': HBOS(contamination=outliers_fraction),
        #         'Isolation Forest': IForest(contamination=outliers_fraction,random_state=random_state),
        #         'K Nearest Neighbors (KNN)': KNN(contamination=outliers_fraction),
        #         'Average KNN': KNN(method='mean',contamination=outliers_fraction)
        # }

        classifiers = {
                'Isolation Forest': IForest(contamination=outliers_fraction,random_state=random_state),
                'Histogram-base Outlier Detection (HBOS)': HBOS(contamination=outliers_fraction),
                'Cluster-based Local Outlier Factor (CBLOF)':CBLOF(contamination=outliers_fraction,check_estimator=False, random_state=random_state),
        }


        #fitting
        # f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')

        # ax = [ax1, ax2, ax3, ax4]

        xx , yy = np.meshgrid(np.linspace(0,1 , 200), np.linspace(0, 1, 200))
        legB = 1
        plt.figure(figsize=(10, 10))
        plt.rcParams.update({'font.size': 20})
        for i, (clf_name, clf) in enumerate(classifiers.items()):


            clf.fit(X)
            # predict raw anomaly score
            scores_pred = clf.decision_function(X) * -1

#             # prediction of a datapoint category outlier or inlier
            y_pred = clf.predict(X)
            n_inliers = len(y_pred) - np.count_nonzero(y_pred)
            n_outliers = np.count_nonzero(y_pred == 1)


            outliers[legA+str(legB)] = clf.predict(X).tolist()
            leg = leg.append(pd.Series([legA+str(legB), cola, colb, clf_name, n_outliers, n_inliers, round(n_outliers/(n_inliers+n_outliers)*100,2)], index=leg.columns ), ignore_index=True)
            #print(outliers.head())
            legB += 1
            
        sums[cola.replace('resource_usage_','')+ ' & '+ colb.replace('resource_usage_','')] = outliers.iloc[:, -3:].sum(axis = 1).tolist()
            
        legA = chr(ord(legA) + 1)
        #print(leg)
        
sums.to_csv('Good_Outputs/sums_three.csv')
outliers.to_csv('Good_Outputs/outliers_three.csv')
leg.to_csv('Good_Outputs/legend_three.csv')
```

```python
#Graph four unsuitable classifiers:

cols = list(usage)
for i in range(len(cols)):
    cola = cols[i]
    for j in range(i+1, len(cols)):
        colb  = cols[j]

# cola = 'resource_usage_bufugu'
# colb = 'resource_usage_business'

        #data prep
        df2 = usage[[cola, colb]].copy()

        X1 = df2[cola].values.reshape(-1,1)
        X2 = df2[colb].values.reshape(-1,1)

        X = np.concatenate((X1,X2),axis=1)


        #model defining
        random_state = np.random.RandomState(42)
        outliers_fraction = 0.05
        # Define seven outlier detection tools to be compared
        classifiers = {
                'Angle-based Outlier Detector (ABOD)': ABOD(contamination=outliers_fraction),
#                 'Cluster-based Local Outlier Factor (CBLOF)':CBLOF(contamination=outliers_fraction,check_estimator=False, random_state=random_state),
                'Feature Bagging':FeatureBagging(LOF(n_neighbors=35),contamination=outliers_fraction,check_estimator=False,random_state=random_state),
#                 'Histogram-base Outlier Detection (HBOS)': HBOS(contamination=outliers_fraction),
#                 'Isolation Forest': IForest(contamination=outliers_fraction,random_state=random_state),
                'K Nearest Neighbors (KNN)': KNN(contamination=outliers_fraction),
                'Average KNN': KNN(method='mean',contamination=outliers_fraction)
        }

#         classifiers = {
#                 'Isolation Forest': IForest(contamination=outliers_fraction,random_state=random_state),
#                 'Histogram-base Outlier Detection (HBOS)': HBOS(contamination=outliers_fraction),
#                 'Cluster-based Local Outlier Factor (CBLOF)':CBLOF(contamination=outliers_fraction,check_estimator=False, random_state=random_state),
#                 'K Nearest Neighbors (KNN)': KNN(contamination=outliers_fraction)
#         }


        #fitting
        # f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')

        # ax = [ax1, ax2, ax3, ax4]

        xx , yy = np.meshgrid(np.linspace(0,1 , 200), np.linspace(0, 1, 200))
        num = 1
        plt.figure(figsize=(10, 10))
        plt.rcParams.update({'font.size': 20})
        for i, (clf_name, clf) in enumerate(classifiers.items()):
        #     ax = plts[0]
            plt.subplot(2, 2, num)
            clf.fit(X)
            # predict raw anomaly score
            scores_pred = clf.decision_function(X) * -1

            # prediction of a datapoint category outlier or inlier
            y_pred = clf.predict(X)
            n_inliers = len(y_pred) - np.count_nonzero(y_pred)
            n_outliers = np.count_nonzero(y_pred == 1)


            # copy of dataframe
            dfx = df2
            dfx['outlier'] = y_pred.tolist()

            # IX1 - inlier feature 1,  IX2 - inlier feature 2
            IX1 =  np.array(dfx[cola][dfx['outlier'] == 0]).reshape(-1,1)
            IX2 =  np.array(dfx[colb][dfx['outlier'] == 0]).reshape(-1,1)

            # OX1 - outlier feature 1, OX2 - outlier feature 2
            OX1 =  dfx[cola][dfx['outlier'] == 1].values.reshape(-1,1)
            OX2 =  dfx[colb][dfx['outlier'] == 1].values.reshape(-1,1)

            print('OUTLIERS : ',n_outliers,'INLIERS : ',n_inliers, clf_name)

            # threshold value to consider a datapoint inlier or outlier
            threshold = stats.scoreatpercentile(scores_pred,100 * outliers_fraction)

            # decision function calculates the raw anomaly score for every point
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]) * -1
            Z = Z.reshape(xx.shape)

            # fill blue map colormap from minimum anomaly score to threshold value
            try:
                plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), threshold, 7),cmap=plt.cm.Blues_r)
            except:
                pass
            # draw red contour line where anomaly score is equal to thresold
            try:
                a = plt.contour(xx, yy, Z, levels=[threshold],linewidths=2, colors='red')
            except:
                pass
            # fill orange contour lines where range of anomaly score is from threshold to maximum anomaly score
            try:
                plt.contourf(xx, yy, Z, levels=[threshold, Z.max()],colors='orange')
            except:
                pass

            b = plt.scatter(IX1,IX2, c='white',s=10, edgecolor='k')

            c = plt.scatter(OX1,OX2, c='black',s=10, edgecolor='k')

            plt.axis('tight') 




#             plt.legend(
#             [a.collections[0], b,c],
#             ['learned decision function', 'inliers: '+str(n_inliers),'outliers: '+ str(n_outliers)],
#             loc='best', facecolor = 'white', frameon = 1, handlelength=0)
            
            plt.legend(
            [b,c,a.collections[0]],
            ['inliers: '+str(n_inliers),'outliers: '+ str(n_outliers),'split: '+str(round(n_outliers/(n_inliers+n_outliers)*100,2))+'%'],
            loc='best', facecolor = 'white', frameon = 1, handlelength=0)

            plt.xlim((0, df2[cola].max()))
            plt.ylim((0, df2[colb].max()))
            plt.title(clf_name)
            num += 1
#         plt.figlegend(
#                 [a.collections[0], b,c],
#                 ['learned decision function', 'inliers','outliers'],
#                 prop=matplotlib.font_manager.FontProperties(size=20),
#                 loc=8,ncol=3)
#         plt.suptitle(cola.replace('resource_usage_','')+ ' & '+ colb.replace('resource_usage_',''))
        plt.tight_layout()
        plt.savefig('unsuitable_graphs/'+cola.replace('resource_usage_','')+ ' & '+ colb.replace('resource_usage_','')+'.png', dpi=150, bbox_inches='tight')
        print(cola+ ' & ' +colb)
        #plt.show()


```
